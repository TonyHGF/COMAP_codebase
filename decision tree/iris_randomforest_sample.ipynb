```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
X = iris.data
y = iris.target
print(type(X))
print(type(y))
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.37,random_state=42)
```

```python
single_tree = dtc(max_depth=1, min_samples_split=5)
single_tree.fit(Xtrain,ytrain)
print(single_tree.score(Xtrain,ytrain))
print(single_tree.score(Xtest,ytest))
```

```python
forest = rfc(n_estimators=6, max_depth=1)
forest.fit(Xtrain, ytrain)
print(forest.score(Xtrain, ytrain))
print(forest.score(Xtest, ytest))
```

```python
for id, tree in enumerate(forest.estimators_):
    plt.figure(figsize=(5,5))      #设置画布大小（单位为英寸）
    sklearn.tree.plot_tree(tree             #训练好的决策树评估器
                          ,node_ids=True    #显示节点id
                          ,filled=True      #给节点填充颜色
                          ,rounded=True     #节点方框变成圆角
                          ,fontsize=12      #节点中文本的字体大小
                          )

    plt.savefig(f"随机森林{id}.png")      #保存图片一定要在plt.show()之前
    plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier as bc

forest2 = bc(n_estimators=6, estimator=dtc(max_depth=1))
forest2.fit(Xtrain, ytrain)

print(forest2.score(Xtrain, ytrain))
print(forest2.score(Xtest, ytest))
```
